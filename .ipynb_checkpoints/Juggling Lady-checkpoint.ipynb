# try to track each ball seperetly:

For an estimation of velocity you use v=dx/dt, where dx is the difference in position (it might have 3 components (x,y,z)) and dt is the difference in time (time between the 2 consecutive frames). For acceleration you need 3 frames as a=dv/dt.

# methods required:

```python
def oneCtorVarifier(contour,mask):
    
    kernel = np.ones((5,5), np.uint8)
    
    if len(contour)==1 :
        mask = cv2.dilate(mask, kernel, iterations=1)
        
    elif len(contour)>1:
        mask = cv2.erode(mask,kernel,iterations=2)
```

```python
def eucledeanDistance(X_old, Y_old, X_new, Y_new):
    eDistance = 0
    if (X_old and Y_old) != 0:
        eDistance = math.dist([X_new,Y_new], [X_old,Y_old])
    return eDistance   
```

```python
def centerCalc(contour):
    
    M = cv2.moments(contour)
    if M["m00"]!= 0:
        cX = int(M["m10"] / M["m00"])
        cY = int(M["m01"] / M["m00"])
        return cX,cY
    else:
        return None
```

# Green Ball Track and speed

```python
cap = cv2.VideoCapture('P4_juggling.mov')
ret, F0 = cap.read()
# fps = cap.get(cv2.CAP_PROP_FPS)
eDistance = 0
lower_g = np.array([40,100,100])
upper_g = np.array([60,255,255])


kernel = np.ones((5,5), np.uint8)
X_g_old=0
Y_g_old=0
              
while(1):
    ret ,F1 = cap.read()
    fps = cap.get(cv2.CAP_PROP_FPS)
    if ret == True:
        F1 = cv2.resize(F1, (640, 360)) 
        blur = cv2.GaussianBlur(F1 ,(0,0),5)
        hsv = cv2.cvtColor(blur, cv2.COLOR_BGR2HSV)    
        mask_green = cv2.inRange(hsv,lower_g,upper_g)

        
        contours_green, _ = cv2.findContours(mask_green, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
        
        #make sure that theres only one contour
        if len(contours_green) == 1:
            mask_green = cv2.dilate(mask_green, kernel, iterations=1)            
        elif len(contours_green) > 1:          
            mask_green = cv2.erode(mask_green,kernel,iterations=2)
        
        
        if(centerCalc(contours_green[0]) != None):
            X_g,Y_g=centerCalc(contours_green[0])

             
        cv2.circle(mask_green, (X_g, Y_g), 1, (0, 0, 0), 2)
        
        cv2.circle(mask_green, (X_g, Y_g), 11, (210, 255, 255), 2)      

        
        #difference between two concecutive frames (Position)
        if (X_g_old and Y_g_old)!=0:
            eDistance = math.dist([X_g,Y_g], [X_g_old,Y_g_old])
        
        timebetweenframes= 1/fps
        velocity = eDistance/timebetweenframes
        X_g_old = X_g
        Y_g_old = Y_g
         
        F0=F1
        
        cv2.putText(mask_green, "{0}".format(math.trunc(velocity)), (X_g - 20, Y_g - 20),
          cv2.FONT_ITALIC, 0.5, (255, 255, 255), 2)
        
        cv2.imshow('final',mask_green) # display frame in a window
        
        k = cv2.waitKey(50) & 0xff
        if k == 27:
            break
    else:
        break


cap.release() # release input video
cv2.destroyAllWindows() # delete output window
cv2.waitKey(1);
```

# Blue Ball Track and speed

```python
cap = cv2.VideoCapture('P4_juggling.mov')
ret, F0 = cap.read()
# fps = cap.get(cv2.CAP_PROP_FPS)

lower_b = np.array([100,60,110])
upper_b = np.array([130,255,255])
eDistance=0

kernel = np.ones((5,5), np.uint8)
X_b_old=0
Y_b_old=0
              
while(1):
    ret ,F1 = cap.read()
    fps = cap.get(cv2.CAP_PROP_FPS)
    if ret == True:
        F1 = cv2.resize(F1, (640, 360)) 
        blur = cv2.GaussianBlur(F1 ,(0,0),5)
        hsv = cv2.cvtColor(blur, cv2.COLOR_BGR2HSV)    
        mask_blue = cv2.inRange(hsv,lower_b,upper_b)

        
        contours_blue, _ = cv2.findContours(mask_blue, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
        
#         make sure that theres only one contour
        if len(contours_blue)==1 :
            mask_blue = cv2.dilate(mask_blue, kernel, iterations=1)
        elif len(contours_blue)>1:
            mask_blue = cv2.erode(mask_blue,kernel,iterations=2)
        
        
        if(centerCalc(contours_blue[0])!=None):
            X_b,Y_b=centerCalc(contours_blue[0])

        cv2.circle(F1, (X_b, Y_b), 1, (0, 0, 0), 2)  # Center       
        cv2.circle(F1, (X_b, Y_b), 11, (200, 255,255), 2) #  Ball Perimeter
      
        
        #difference between two concecutive frames (Position)
        if (X_b_old and Y_b_old)!=0:
            eDistance = math.dist([X_b,Y_b], [X_b_old,Y_b_old])
        
        #difference between two concecutive frames (time)
        timebetweenframes= 1/fps
        
        
        velocity = eDistance/timebetweenframes
        X_b_old = X_b
        Y_b_old = Y_b
         
        F0=F1
        
        cv2.putText(F1, "{0}".format(math.trunc(velocity)), (X_b - 20, Y_b - 20),
          cv2.FONT_ITALIC, 0.5, (255, 255, 255), 2)
        
        cv2.imshow('final',F1) # display frame in a window
#         print(1/fps)
        
        k = cv2.waitKey(80) & 0xff
        if k == 27:
            break
    else:
        break


cap.release() # release input video
cv2.destroyAllWindows() # delete output window
cv2.waitKey(1);
```

# All Together

```python
cap = cv2.VideoCapture('P4_juggling.mov')

W = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
H = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
timebetweenframes= 1/fps   #difference between two concecutive frames (time)

fourcc = cv2.VideoWriter_fourcc(*'.mov') # or just copy it from the original video

out = cv2.VideoWriter('Final_Juggling_Lady.mov',fourcc, fps, (W,H)) # (W,H) should be integers

ret, F0 = cap.read()

# thresholds for each ball
lower_g = np.array([40,100,100])
upper_g = np.array([60,255,255])

lower_b = np.array([100,60,110])
upper_b = np.array([130,255,255])

lower_r = np.array([1,100,184])
upper_r = np.array([5,180,250])


# initiating coordinates
X_b_old, X_g_old, X_r_old=0,0,0
Y_b_old, Y_g_old, Y_r_old=0,0,0
 
lineX1 = 250
lineX2 = 420
counter = 0
flag = False

cv2.namedWindow('final')

cv2.createTrackbar('Thickness','final',2,10,(lambda a: None))
              
while(1):
    ret ,F1 = cap.read()
    
    if ret == True:
        
        
        F1 = cv2.resize(F1, (640, 360)) 
       
        blur = cv2.GaussianBlur(F1 ,(0,0),5)
        hsv = cv2.cvtColor(blur, cv2.COLOR_BGR2HSV)    
        mask_blue = cv2.inRange(hsv,lower_b,upper_b)
        mask_red = cv2.inRange(hsv,lower_r,upper_r)
        mask_green = cv2.inRange(hsv,lower_g,upper_g)
        
         
        cv2.rectangle(F1, (lineX1, 0), (lineX1, H), (0, 255, 0), 2)  
        cv2.rectangle(F1, (lineX2, 0), (lineX2, H), (0, 255, 0), 2 )   
        
        output = F1.copy()
        
        contours_blue, _ = cv2.findContours(mask_blue, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
        contours_red, _ = cv2.findContours(mask_red, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
        contours_green, _ = cv2.findContours(mask_green, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
        
        # make sure that theres only one contour
        oneCtorVarifier(contours_blue, mask_blue)
        oneCtorVarifier(contours_red, mask_red)
        oneCtorVarifier(contours_green, mask_green)
        
        
        if(centerCalc(contours_blue[0])!=None):
            X_b, Y_b = centerCalc(contours_blue[0])
            
        if(centerCalc(contours_red[0])!=None):
            X_r, Y_r = centerCalc(contours_red[0])
            
        if(centerCalc(contours_green[0])!=None):
            X_g, Y_g = centerCalc(contours_green[0])
             
        
        #difference between two concecutive frames (Position)
        eDistance_blue = eucledeanDistance(X_b_old,Y_b_old,X_b,Y_b)
        eDistance_green = eucledeanDistance(X_g_old,Y_g_old,X_g,Y_g)
        eDistance_red = eucledeanDistance(X_r_old,Y_r_old,X_r,Y_r)
        
        #distance between balls
        dist_red2blue = eucledeanDistance(X_b,Y_b,X_r,Y_r)
        dist_red2green = eucledeanDistance(X_g,Y_g,X_r,Y_r)
        dist_green2blue = eucledeanDistance(X_b,Y_b,X_g,Y_g)
        
        # v = dx/dt
        velocity_blue = eDistance_blue/timebetweenframes # v = dx/dt
        velocity_green = eDistance_green/timebetweenframes
        velocity_red = eDistance_red/timebetweenframes
        
        # each iteration the current coordinates will become the old ones inorder to determine position difference from frame to frame
        X_b_old = X_b
        Y_b_old = Y_b
        X_g_old = X_g
        Y_g_old = Y_g
        X_r_old = X_r
        Y_r_old = Y_r
         
                
        thickness = cv2.getTrackbarPos('Thickness','final')
        
        # Drawing speed of each ball
        cv2.rectangle(F1, (X_r + 20, Y_r - 15), (X_r + 55, Y_r - 22), (0,0,0), 12)
        cv2.rectangle(F1, (X_b + 20, Y_b - 15), (X_b + 55, Y_b - 22), (0,0,0), 12)
        cv2.rectangle(F1, (X_g + 20, Y_g - 15), (X_g + 55, Y_g - 22), (0,0,0), 12)
        
        cv2.addWeighted(F1,  0.3, output, 1 - 0.3, 0, output)
        
         # To check if any ball got past the vertical lines
        if lineX1 > X_b or lineX1 > X_g or lineX1 > X_r:     
            cv2.line(output, (lineX1, 0), (lineX1, H), (0, 0, 255), 3 )
            if flag == False:
                flag = True
                counter += 1
        elif lineX2 < X_b or lineX2 < X_g or lineX2 < X_r: 
            cv2.line(output, (lineX2, 0), (lineX2, H), (0, 0, 255), 3 )
            if flag == False:
                flag = True
                counter += 1
        else:
            flag = False
           
            
        #text of speed frame of every balls per frame
        cv2.putText(output, "{0}".format(math.trunc(velocity_blue)), (X_b + 15, Y_b - 15),
          cv2.FONT_ITALIC, 0.5, (255, 255, 255), 2)
        cv2.putText(output, "{0}".format(math.trunc(velocity_green)), (X_g + 15, Y_g - 15),
          cv2.FONT_ITALIC, 0.5, (255, 255, 255), 2)
        cv2.putText(output, "{0}".format(math.trunc(velocity_red)), (X_r + 15, Y_r - 15),
          cv2.FONT_ITALIC, 0.5, (255, 255, 255), 2)
        
        # sign line between balls
        cv2.line(output, (X_b,Y_b), (X_r,Y_r), (0,0,255), 2)
        cv2.line(output, (X_g,Y_g), (X_r,Y_r), (0,0,255), 2)
        cv2.line(output, (X_b,Y_b), (X_g,Y_g), (0,0,255), 2)
        
        #text of distance between balls
        cv2.putText(output, "Distance Green 2 Blue: {0}".format(math.trunc(dist_green2blue)), ( 10, 15),
          cv2.FONT_ITALIC, 0.5, (255, 255, 255), 2)
        cv2.putText(output, "Distance Red 2 Blue: {0}".format(math.trunc(dist_red2blue)), ( 10, 35),
          cv2.FONT_ITALIC, 0.5, (255, 255, 255), 2)
        cv2.putText(output, "Distance Green 2 Red: {0}".format(math.trunc(dist_red2green)), ( 10, 55),
          cv2.FONT_ITALIC, 0.5, (255, 255, 255), 2)
        cv2.putText(output, "Counter: {0}".format(math.trunc(counter)), ( 10, 75),
          cv2.FONT_ITALIC, 0.5, (255, 255, 255), 2)
        
                  
        cv2.circle(output, (X_b, Y_b), 10, (230, 255,255), thickness) #   blue Ball Perimeter           
        cv2.circle(output, (X_r, Y_r),  10, (230, 255, 255), thickness)#   red ball Perimeter             
        cv2.circle(output, (X_g, Y_g), 10, (230, 255, 255), thickness)#   green ball Perimeter     
            
        cv2.imshow('final',output) # display frame in a window
        out.write(output)
        

        k = cv2.waitKey(20) & 0xff
        if k == 27:
            break
    else:
        break

cap.release() # release input video
cv2.destroyAllWindows() # delete output window
cv2.waitKey(1);
```
